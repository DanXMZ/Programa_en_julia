# Introduccion a Makie.jl

*Makie* es un ecosistema de visualización de datos para el lenguaje de programación Julia , con alto rendimiento y extensibilidad. *Makie* es el nombre de todo el ecosistema de trazado y Makie.jl es el paquete principal que describe cómo funcionan los trazados.
    Para trabajar con los gráficos necesitamos un *backend* que los traduzca en imagenes o gráficos vectoriales. Existen tres  principalmente : *CairoMakie.jl*, *GLMakie* y *WGLMakie*, estos ultimos con la posibilidad de trabajar en imagenes 3D y ventanas interactivas.

### Definiendo un objeto contenedor básico

Los tres elementos básicos de un contenedor en *Makie* son **figure**, **axis** y **plots**. Por lo general en un gráfico de *Makie* encontraras un *figure* el cual contiene un *axis* (eje) que contienen uno o mas objetos que formaran wl o los gráficos, *lines* o *scatter*.

```julia
using GLMakie
f = Figure(backgroundcolor = :blue,resolution=(800,300))#Creamos una figure y definimos color y tamaño
```

### Definiendo un eje

```julia
f = Figure(resolution = (400,400))
ax = Axis(f[1, 1])
f
```

### Nombrar ejes

```julia
f = Figure(resolution = (300,300))
ax = Axis(f[1, 1],
    title = "Ejes de Makie",
    xlabel = "etiqueta eje x",
    ylabel = "etiqueta eje y"
)
f
```

### Usando función *lines!*

```julia
f = Figure(resolution =(500,500))
ax = Axis(f[1, 1])
x = range(0, 10, length=100)
y = sin.(x)
lines!(x, y)
f
```

### Ejemplo de uso de función *scatter!*

```julia
f = Figure(resolution =(500,500))
ax = Axis(f[1, 1])
x = range(0, 15, length=100)
y = sin.(x)
scatter!(x, y)
f
```

Hemos creado los gráficos definiendo estos tres elementos, sin embargo, existe la posibilidad de crear dichos gráficos de forma que implicitamente se definan estos tres, 

```julia
x = range(0, 15, length=50)
y = sin.(x)
lines(x, y)
```

```julia
x = range(0, 10, length=100)
y = sin.(x)
scatter(x, y)
```

```julia
figure, axis, lineplot = scatter(x, y)
figure
```

### 

```julia
x = range(0, 10, length=100)
y = sin.(x)
scatter(x, y;
    figure = (; resolution = (800, 600)),
    axis = (; title = "Scatter plot", xlabel = "x label")
)
```

```julia
function mandelbrot(a)
    z = 0
    for i=1:50
        z = z^2 + a
    end
    return z
end

for y=1.0:-0.05:-1.0
    for x=-2.0:0.0315:0.5
        abs(mandelbrot(complex(x, y))) < 2 ? print("*") : print(" ")
    end
    println()
end
```
